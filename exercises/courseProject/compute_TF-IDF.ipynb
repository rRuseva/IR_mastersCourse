```python
### for a given document (title or caption) computes the term frequency 
### returns dictionary with all terms from all documents for keys and TF for the values
def computeTFperDoc(doc, bow):
    termsDict = dict.fromkeys(bow, 0)
    n = len(doc)
    for term in doc:
        termsDict[term] = doc.count(term)/n
    return termsDict

termsDict = computeTFperDoc(rp.cleanedDocs[0]['caption'][0], bow)
# print(len(termsDict))
# print(termsDict)
df1 = pd.DataFrame.from_dict(termsDict, orient='index')
df1.sort_values(by=[0], ascending=False)
```

```python
### TF - the number of times a term appears in a document/ devided by the total number of terms in that document

def computeTF(corpus,bow):
    
#     bagOfWordsTitle, bagOfWordsCaption = createBagsOfWOrds(rp.cleanedDocs)
#     bow = set(bagOfWordsTitle).union(set(bagOfWordsCaption))
    title_tf = []
    caption_tf = []
    for doc in corpus:
        title_tf.append(computeTFperDoc(doc['articleTitle'], bow))
        for cap in doc['caption']:
            caption_tf.append(computeTFperDoc(cap, bow))
    return title_tf, caption_tf
    
title_tf, caption_tf = computeTF(rp.cleanedDocs,bow)
print(len(title_tf))
print(len(caption_tf))

# print(title_tf[0]['case'])
# print(title_tf[1]['case'])
# print(title_tf[2]['case'])
# print(title_tf[3]['case'])
# print(title_tf[4])
```

```python
### computes the idf for all the terms in the 'Bag of words' from all docs
import math
def computeIDF(corpus, bow):
    titleN = len(corpus)
    captionN = 0
#     bagOfWordsTitle, bagOfWordsCaption = createBagsOfWOrds(rp.cleanedDocs)
#     bow = set(bagOfWordsTitle).union(set(bagOfWordsCaption))
    
    title_idfDict = dict.fromkeys(bow,[])
    caption_idfDict = dict.fromkeys(bow,[])
    for i,doc in enumerate(corpus):
        for term in doc['articleTitle']:
            if len(title_idfDict[term]) > 0:  ### term in doc['articleTitle'] :
                title_idfDict[term].append((i,'t')) 
            else:
                title_idfDict[term] = [(i,'t')]
        for j, cap in enumerate(doc['caption']):
            captionN += 1
            for term in cap:
                if len(caption_idfDict[term]) > 0:
                    caption_idfDict[term].append((i,j,'c'))
                else:
                    caption_idfDict[term] = [(i,j,'c')]
                    
    for term in title_idfDict:
        title_idfDict[term] = math.log10(titleN/(len(title_idfDict[term])+1))
    for term in caption_idfDict:
        caption_idfDict[term] = math.log10(captionN/(len(caption_idfDict[term])+1))
        
    return title_idfDict, caption_idfDict

title_idfDict, caption_idfDict = computeIDF(rp.cleanedDocs, bow)
# print(len(title_idfDict))
# print(title_idfDict)
# print(caption_idfDict['case'])
```

```python
### computes the TF-IDF weight by given tfs and idfs for the terms in all docs (titles and captions)
def computeTFIDF(tfs,idfs, bow):
#     bagOfWordsTitle, bagOfWordsCaption = createBagsOfWOrds(rp.cleanedDocs)
#     bow = set(bagOfWordsTitle).union(set(bagOfWordsCaption))
    
    tfidf_dict = [] 
    n = len(tfs)
    
    for i in range(n):
        tfidf_doc = dict.fromkeys(bow,0)
        for term in idfs.keys():
#              print(term, tfs[i][term], idfs[term])
            tfidf_doc[term] = tfs[i][term]*idfs[term]
        tfidf_dict.append(tfidf_doc)
    return tfidf_dict

title_tfidf = computeTFIDF(title_tf, title_idfDict, bow)
df = pd.DataFrame(list(title_tfidf))
# df
```

```python
caption_tfidf = computeTFIDF(caption_tf, caption_idfDict,bow)
df = pd.DataFrame(list(caption_tfidf))
# df
# caption_tfidf[0]
```

```python
### computes the final TF-IDF weijght for each term by the formula:
### TF-IDF weight = alpha * title_tfidf + (1-alpha)caption_tfidf

def mergeTFIDFtables(alpha):
    tfidf = []
    title_tfidf = computeTFIDF(title_tf, title_idfDict, bow)
    caption_tfidf = computeTFIDF(caption_tf, caption_idfDict, bow)
    captions_dict = rp.capDicts(rp.cleanedDocs)
    n = len(caption_tfidf)
    m = len(title_tfidf)
    doc_idx = 0 
    for i in range(n-1):
        doc_tfidf =  dict.fromkeys(bow,0)
        
        for term in bow:
            doc_tfidf[term] = alpha*caption_tfidf[i][term] + title_tfidf[captions_dict[i]][term]*(1-alpha)
        
        tfidf.append(doc_tfidf)
    
    return tfidf
            
finalTFIDF = mergeTFIDFtables(0.3)

# print(finalTFIDF)
    
print
```
